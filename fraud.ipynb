```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the Dataset

```python
df = pd.read_csv("fraud.csv")
```

### Data Cleaning

```python
df.columns
```

```python
# Removing Unnecessary Columns

df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Converting trans_date_trans_time column

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
```

```python
# Extracting new useful columns

df['hour'] = df['trans_date_trans_time'].dt.hour
df['day'] = df['trans_date_trans_time'].dt.day
df['month'] = df['trans_date_trans_time'].dt.month
```

### Basic Transaction Statisics

```python
print("Total transaction amount is :", round(df['amt'].sum(),2))
```

```python
print("Average transaction value is :", round(df['amt'].mean(),2))
```

```python
fraud_counts = df['is_fraud'].value_counts()

print(f"Number of non-fraud transactions : {fraud_counts[0]}" )
print(f"Number of fraud transactions : {fraud_counts[1]}")
```

### Fraud Distribution

```python
fraud_counts = df['is_fraud'].value_counts()

plt.figure()
fraud_counts.plot(kind='bar')
plt.title("Fraud vs Non-Fraud Transactions")
plt.ylabel("Count")
plt.show()

print("Insight:\nFraud is usually <1% of transactions")
```

### Transaction Amount Analysis

```python
plt.figure()
df['amt'].hist(bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

### Fraud Rate by Category

```python
fraud_by_category = df.groupby('category')['is_fraud'].sum()

plt.figure()
fraud_by_category.plot(kind='bar')
plt.title("Fraud by Transaction Category")
plt.ylabel("Fraud Count")
plt.show()
```

### Fraud by Gender

```python
fraud_gender = df.groupby('gender')['is_fraud'].sum()

plt.figure()
fraud_gender.plot(kind='bar')
plt.title("Fraud by Gender")
plt.show()

print("Business insight:\nhelps identify customer behavior patterns")
```

### Fraud by Hour of Day

```python
fraud_hour = df.groupby('hour')['is_fraud'].sum()

plt.figure()
fraud_hour.plot(kind='line')
plt.title("Fraud Transactions by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Count")
plt.show()

print("Insight:\nfraud often spikes late night")
```

### Fraud by Age Group

```python
# Age calculation

df['dob'] = pd.to_datetime(df['dob'])
df['age'] = df['trans_date_trans_time'].dt.year-df['dob'].dt.year
```

```python
df['age_group'] = pd.cut(df['age'], bins=[18,30,40,50,60,70,90])

fraud_age = df.groupby('age_group')['is_fraud'].sum()

plt.figure()
fraud_age.plot(kind='bar')
plt.title("Fraud by Age Group")
plt.show()

print("Insight:\nidentifies riskier customer segments")
```

### Distance Between Customer and Merchant

```python
df['distance'] = ((df['lat'] - df['merch_lat'])**2 +
                  (df['long'] - df['merch_long'])**2)**0.5

plt.figure()
df['distance'].hist(bins=50)
plt.title("Customer-Merchant Distance")
plt.show()

print("Insight:\nfraud often occurs when distance is unusually large")
```

### Fraud Count By State

```python
df[df['is_fraud']==1].groupby('state')['is_fraud'].count().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Fraud Transactions by State")
plt.xlabel("State")
plt.ylabel("Fraud Count")
plt.show()
```
